```python
import numpy as np	
import cv2
from matplotlib import pyplot as plt

imagen = cv2.imread('original.jpg')
img_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
cv2.imwrite('imag_gris.jpg', img_gris)
plt.imshow(img_gris, cmap = "gray");
```

HISTOGRAMA COLOR

```python
def Histograma_Comp_color(Image):
	Image_Height = Image.shape[0]
	Image_Width = Image.shape[1]
	Image_Channels = Image.shape[2]
	
	Histograma = np.zeros([256, Image_Channels], np.int32)
	
	for x in range(0, Image_Height):
		for y in range(0, Image_Width):
			for c in range(0, Image_Channels):
					Histograma[Image[x,y,c], c] +=1
	
	return Histograma

def Plot_Histograma_color(Histograma):
	plt.figure()
	plt.title("Histograma - Color")
	plt.xlabel("Intensidad")
	plt.ylabel("Frecuencia de Intensidad")
	plt.xlim([0, 256])
	plt.plot(Histograma[:,0],'b') # B
	plt.plot(Histograma[:,1],'g') # G
	plt.plot(Histograma[:,2],'r') # R
	plt.savefig("Color_Histograma.jpg")
```

```python
Histograma = Histograma_Comp_color(imagen)

for i in range(0, Histograma.shape[0]):
	for c in range(0, Histograma.shape[1]):
		print("Histograma[", i, ", ", c, "]: ", Histograma[i,c])
		
Plot_Histograma_color(Histograma)
```

Histograma escala de grises

```python

def HistogramaComp_gray(Image):
	Image_Height = Image.shape[0]
	Image_Width = Image.shape[1]
	Histograma = np.zeros([256], np.int32)
	for x in range(0, Image_Height):
		for y in range(0, Image_Width):
			Histograma[Image[x,y]] +=1
	return Histograma

def Plot_Histograma_gray(Histograma):
	plt.figure()
	plt.title("Histograma - Escala de Grises")
	plt.xlabel("Intensidad")
	plt.ylabel("Intensity Frequency")
	plt.xlim([0, 256])
	plt.plot(Histograma)
	plt.savefig("Histograma_GrayScale.jpg")
```

```python
Histograma_GrayScale = HistogramaComp_gray(img_gris)
for i in range(0,len(Histograma_GrayScale)):
		print("Histograma[",i,"]: ", Histograma_GrayScale[i])
Plot_Histograma_gray(Histograma_GrayScale)
```

```python
alpha = 2  # Factor de contraste
beta = 0     # Factor de brillo
imagen_contraste = cv2.convertScaleAbs(img_gris, alpha=alpha, beta=beta)
imagen_ecualizada = cv2.equalizeHist(img_gris)

hist_contraste = HistogramaComp_gray(imagen_contraste)
hist_ecualizado = HistogramaComp_gray(imagen_ecualizada)
```

```python
def mostrar_comparacion():
    plt.figure(figsize=(15, 10))
        # Imágenes
    plt.subplot(231), plt.imshow(img_gris, cmap='gray')
    plt.title('Imagen Original')
    plt.subplot(232), plt.imshow(imagen_contraste, cmap='gray')
    plt.title('Aumento de Contraste')
    plt.subplot(233), plt.imshow(imagen_ecualizada, cmap='gray')
    plt.title('Ecualización del Histograma')
        # Histogramas
    plt.subplot(234), plt.plot(Histograma_GrayScale)
    plt.title('Histograma Original')
    plt.subplot(235), plt.plot(hist_contraste)
    plt.savefig('Histograma_GrayScale.jpg')
    plt.title('Histograma Contraste')
    plt.subplot(236), plt.plot(hist_ecualizado)
    plt.title('Histograma Ecualizado')
    plt.tight_layout()
    plt.savefig('comparacion_completa.jpg')
    plt.show()

    
cv2.imwrite('imagen_contraste.jpg', imagen_contraste)
cv2.imwrite('imagen_ecualizada.jpg', imagen_ecualizada)
mostrar_comparacion()
```

```python
def mostrar_imagenes():
    plt.figure(figsize=(20, 10))
    
    plt.subplot(241), plt.imshow(img_gris, cmap='gray'), plt.title('Escala de Grises')
    plt.subplot(242), plt.imshow(img_ruidosa, cmap='gray'), plt.title('Con Ruido Gaussiano')
    plt.subplot(243), plt.imshow(img_suavizada, cmap='gray'), plt.title('Suavizada')
    plt.subplot(244), plt.imshow(img_binaria, cmap='gray'), plt.title('Binaria')
    plt.subplot(245), plt.imshow(img_dilatada, cmap='gray'), plt.title('Dilatada')
    plt.subplot(246), plt.imshow(img_erosionada, cmap='gray'), plt.title('Erosionada')
    plt.savefig('comparacion_procesado1.jpg')
    plt.subplot(247), plt.imshow(bordes_sobel, cmap='gray'), plt.title('Bordes (Umbralización)')
    plt.subplot(248), plt.imshow(bordes_canny, cmap='gray'), plt.title('Bordes (Canny)')
    plt.tight_layout()
    plt.savefig('comparacion_procesado2.jpg')
    plt.show()
```

```python
def aplicar_ruido_gaussiano(imagen, media=0, desviacion=25):
    ruido = np.random.normal(media, desviacion, imagen.shape).astype(np.uint8)
    imagen_ruidosa = cv2.add(imagen, ruido)
    return imagen_ruidosa


img_ruidosa = aplicar_ruido_gaussiano(img_gris)
cv2.imwrite('imagen_ruidosa.jpg', img_ruidosa)
kernel_size = 7
img_suavizada = cv2.GaussianBlur(img_ruidosa, (kernel_size, kernel_size), 0)
cv2.imwrite('imagen_suavizada.jpg', img_suavizada)
_, img_binaria = cv2.threshold(img_gris, 127, 255, cv2.THRESH_BINARY)
cv2.imwrite('imagen_binaria.jpg', img_binaria)
kernel = np.ones((5,5), np.uint8)
img_dilatada = cv2.dilate(img_binaria, kernel, iterations=1)
img_erosionada = cv2.erode(img_binaria, kernel, iterations=1)
cv2.imwrite('imagen_dilatada.jpg', img_dilatada)
cv2.imwrite('imagen_erosionada.jpg', img_erosionada)
_, img_umbralizada = cv2.threshold(img_gris, 170, 255, cv2.THRESH_BINARY)
bordes_sobel = cv2.Sobel(img_gris, cv2.CV_64F, 1, 1, ksize=5)
bordes_sobel = np.uint8(np.absolute(bordes_sobel))
cv2.imwrite('bordes_umbralización.jpg', bordes_sobel)
bordes_canny = cv2.Canny(img_gris, 100, 200)
cv2.imwrite('bordes_canny.jpg', bordes_canny)
print("Hecho")
```
